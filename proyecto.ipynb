```python
import geopandas as gpd
import os

# 1. Cargar el mapa vectorial de Chile.
mapa_regiones = gpd.read_file('Regional.shp')

# Reemplaza con el nombre de tu columna de región
columna_nombre_region = 'Region' 

# 2. Iterar sobre cada región
for nombre_region in mapa_regiones[columna_nombre_region].unique():
    region_actual = mapa_regiones[mapa_regiones[columna_nombre_region] == nombre_region]
    
    # Reproyecta a EPSG:4326
    region_actual = region_actual.to_crs(epsg=4326)
    
    # --- Aplica simplificación con una tolerancia condicional ---
    # Comprueba si el nombre de la región coincide con Magallanes
    if nombre_region == 'Región de Magallanes y de la Antártica Chilena':
        # Aplica una alta tolerancia de 0.6
        region_actual['geometry'] = region_actual.simplify(tolerance=0.4)
    else:
        # Aplica la tolerancia estándar de 0.2 al resto de las regiones
        region_actual['geometry'] = region_actual.simplify(tolerance=0.05)
    
    nombre_archivo = nombre_region.replace(' ', '_').replace('.', '') + '.geojson'
    
    region_actual.to_file(os.path.join(os.getcwd(), nombre_archivo), driver='GeoJSON')
    
    print(f'Exportado: {nombre_archivo}')

print('Proceso completado.')
```

```python
import pandas as pd
import geopandas as gpd
import glob
import os

# 1. Crea una lista para guardar cada GeoDataFrame
gdfs = []

# 2. Encuentra todos los archivos .geojson en el directorio actual
geojson_files = glob.glob('*.geojson')

# 3. Itera sobre cada archivo GeoJSON y lo carga en un GeoDataFrame
for file in geojson_files:
    # Lee el archivo GeoJSON
    gdf_temp = gpd.read_file(file)
    # Agrega el GeoDataFrame a la lista
    gdfs.append(gdf_temp)

# 4. Concatena todos los GeoDataFrames de la lista en uno solo
# Usa pd.concat() en lugar de gpd.concat()
combined_gdf = pd.concat(gdfs, ignore_index=True)

# 5. Guarda el GeoDataFrame combinado en un nuevo archivo GeoJSON
combined_gdf.to_file('regiones_combinadas.geojson', driver='GeoJSON')

print('Archivos GeoJSON combinados exitosamente en "regiones_combinadas.geojson".')
```
